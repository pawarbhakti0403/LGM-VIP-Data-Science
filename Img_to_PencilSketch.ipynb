Importing Libraries







```python
import cv2
import matplotlib.pyplot as plt
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=[15,6]     #Changing default value of figure size
```

Reading an image and converting an image from BGR to RGB

```python
img=cv2.imread('lotus.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)     
plt.imshow(img)
plt.axis('off')
plt.grid(False)
plt.title('Original Image',size=20);
```

Converting an image to Grayscale

```python
gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img,cmap='gray')
plt.axis('off')
plt.grid(False)
plt.title('Grayscale Image',size=20);
```

Inverting an Grayscale image

```python
invert_img=cv2.bitwise_not(gray_img)
plt.imshow(invert_img,cmap='gray')
plt.axis('off')
plt.grid(False)
plt.title('Inverted Grayscale Image',size=20);
```

Converting an Inverted image to Blurry image

```python
blur_img=cv2.GaussianBlur(invert_img,(21,21),sigmaX=0,sigmaY=0)
plt.imshow(blur_img,cmap='gray')
plt.axis('off')
plt.grid(False)
plt.title('Blurry Image',size=20);
```

Mixing Grayscale and Blurry image to make Pencil sketch

```python
pencil_sketch=cv2.divide(gray_img,255-blur_img,scale=255)
plt.imshow(pencil_sketch,cmap='gray')
plt.axis('off')
plt.grid(False)
plt.title('Pencil Sketch',size=20);
```
